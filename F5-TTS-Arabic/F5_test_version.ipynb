```python


from IPython.display import Audio
import os

def display_audio_from_path(audio_path: str):
  """
  Displays audio from a given file path in a Colaboratory/Jupyter notebook.

  Args:
    audio_path: The path to the audio file (e.g., .wav, .mp3).
  """
  if not os.path.exists(audio_path):
    print(f"Error: Audio file not found at path: {audio_path}")
    return

  print(f"Displaying audio from: {audio_path}")
  display(Audio(audio_path))

# Display the generated audio
output_file="tests/new4.wav"
display_audio_from_path(output_file)
```
